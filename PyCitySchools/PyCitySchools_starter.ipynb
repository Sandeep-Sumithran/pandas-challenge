```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score

  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#drop unneeded fields
school_data_required = school_data_complete.drop(['gender','Student ID','School ID'], axis=1)
#school_data_required
```

```python

school_names_summary = school_data_required.groupby('school_name')['type'].unique()
school_T = pd.DataFrame({
    "school_name" : school_names_summary.index,
    "type": school_names_summary.values
})
school_T = school_T.rename(columns={"type":"School Type"})
```

```python
students_per_school = school_data_required.groupby('school_name')['student_name'].count()
school_S = pd.DataFrame({
    "school_name": students_per_school.index,
    "student_name": students_per_school.values
})
school_S = school_S.rename(columns={"student_name":"Total Students"})
#school_S
```

```python
budget_per_school = school_data_required.groupby("school_name")['budget'].unique()
school_B = pd.DataFrame({
    "school_name":budget_per_school.index,
    "budget":budget_per_school.values
})
school_B['budget'] = school_B['budget'].str.get(0)
school_B = school_B.rename(columns={"budget":"Total School Budget"})
```

```python
#merge
school_T_S = pd.merge(school_T,school_S,how="inner",on=["school_name","school_name"])
school_T_S_B = pd.merge(school_T_S, school_B,how="inner",on=["school_name","school_name"])
#school_T_S_B
```

```python
#per student
school_T_S_B["Per Student Budget"] = school_T_S_B["Total School Budget"]/school_T_S_B["Total Students"]
#school_T_S_B
```

```python
school_avg_math = school_data_required.groupby("school_name")["maths_score"].mean()
school_AM = pd.DataFrame({
    "school_name": school_avg_math.index,
    "Average Maths Score":school_avg_math.values
})
#merge
school_T_S_B_AM = pd.merge(school_T_S_B,school_AM,how="inner",on=["school_name","school_name"])
#school_T_S_B_AM
school_avg_read = school_data_required.groupby("school_name")["reading_score"].mean()
school_AR = pd.DataFrame({
    "school_name": school_avg_read.index,
    "Average Reading Score":school_avg_read.values
})
#merge
school_T_S_B_AM_AR = pd.merge(school_T_S_B_AM,school_AR,how="inner",on=["school_name","school_name"])
#school_T_S_B_AM_AR
```

```python
#calculate passing maths score percentages
math_pass = school_data_required.loc[school_data_required['maths_score']>=50]
math_pass_school = math_pass.groupby("school_name")["maths_score"].count()
school_MP = pd.DataFrame({
    "school_name":math_pass_school.index,
    "Students Passing Maths":math_pass_school.values
})
school_C_MP = pd.merge(school_T_S_B_AM_AR, school_MP,how="inner", on=["school_name", "school_name"])

school_C_MP["% Passing Maths"] = (school_C_MP["Students Passing Maths"]/school_C_MP["Total Students"])*100
school_C_MP = school_C_MP.drop(["Students Passing Maths"], axis=1)
#school_C_MP
```

```python
#calculate passing reading score percentages
read_pass = school_data_required.loc[school_data_required['reading_score']>=50]
read_pass_school = read_pass.groupby("school_name")["reading_score"].count()
school_RP = pd.DataFrame({
    "school_name":read_pass_school.index,
    "Students Passing Reading":read_pass_school.values
})
school_C_MP_RP = pd.merge(school_C_MP, school_RP,how="inner", on=["school_name", "school_name"])

school_C_MP_RP["% Passing Reading"] = (school_C_MP_RP["Students Passing Reading"]/school_C_MP_RP["Total Students"])*100
school_C_MP_RP = school_C_MP_RP.drop(["Students Passing Reading"], axis=1)
#school_C_MP_RP
```

```python
#calculate overall pass rate
school_data_required["Overall Passing"] = (school_data_required["maths_score"]>=50) &(school_data_required["reading_score"]>=50)
overall_passing = school_data_required.loc[school_data_required["Overall Passing"] == True]
overall_pass_school = overall_passing.groupby("school_name")["Overall Passing"].count()

school_O = pd.DataFrame({
    "school_name": overall_pass_school.index,
    "Students Overall Passing": overall_pass_school.values
})
school_combined = pd.merge(school_C_MP_RP, school_O, how="inner", on=["school_name", "school_name"])
school_combined["% Overall Passing"] = (school_combined["Students Overall Passing"]/school_combined["Total Students"])*100
school_combined = school_combined.drop(["Students Overall Passing"], axis=1)
#school_combined
```

```python
#formatting
per_school_summary = school_combined.set_index("school_name")
per_school_summary["Total Students"] = per_school_summary["Total Students"].map("{:,}".format)
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary['School Type'] = per_school_summary['School Type'].astype(str).str.strip("[]").str.replace("'", "")
per_school_summary.index.name = None

#variables for later
school_size = per_school_summary
school_type_analysis = per_school_summary
school_spending = per_school_summary

per_school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#remove formatting
school_spending["Per Student Budget"] = school_spending["Per Student Budget"].str.replace("$", "", regex=False)
school_spending["Per Student Budget"] = school_spending["Per Student Budget"].astype(float)

# Create bins
bins = [0, 585, 630, 645, 680]
group_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_spending["Spending Ranges (Per Student)"] = pd.cut(school_spending["Per Student Budget"], bins, labels = group_labels, include_lowest = True)

spending_m_scores = school_spending.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["Average Maths Score"].map("{:.2f}".format)
spending_r_scores = school_spending.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["Average Reading Score"].map("{:.2f}".format)
spending_p_m = school_spending.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Passing Maths"].map("{:.2f}".format)
spending_p_r = school_spending.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Passing Reading"].map("{:.2f}".format)
overall_pass_spend = school_spending.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Overall Passing"].map("{:.2f}".format)

spending_summary = pd.DataFrame({
    "Average Maths Score": spending_m_scores,
    "Average Reading Score": spending_r_scores,
    "% Passing Maths": spending_p_m,
    "% Passing Reading": spending_p_r,
    "% Overall Passing": overall_pass_spend
})

spending_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#remove formatting
school_size["Total Students"] = school_size["Total Students"].str.replace(",", "", regex=False)
school_size["Total Students"] = school_size["Total Students"].astype(float)

# Create bins
bins_s = [0, 1000,2000,5000]
group_labels_s = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_size["School Size"] = pd.cut(school_size["Total Students"], bins_s, labels = group_labels_s, include_lowest = True)

size_m_scores = school_size.groupby(["School Size"]).mean(numeric_only=True)["Average Maths Score"].map("{:.2f}".format)
size_r_scores = school_size.groupby(["School Size"]).mean(numeric_only=True)["Average Reading Score"].map("{:.2f}".format)
size_p_m = school_size.groupby(["School Size"]).mean(numeric_only=True)["% Passing Maths"].map("{:.2f}".format)
size_p_r = school_size.groupby(["School Size"]).mean(numeric_only=True)["% Passing Reading"].map("{:.2f}".format)
overall_pass_size = school_size.groupby(["School Size"]).mean(numeric_only=True)["% Overall Passing"].map("{:.2f}".format)

size_summary = pd.DataFrame({
    "Average Maths Score": size_m_scores,
    "Average Reading Score": size_r_scores,
    "% Passing Maths": size_p_m,
    "% Passing Reading": size_p_r,
    "% Overall Passing": overall_pass_size
})

size_summary
```
